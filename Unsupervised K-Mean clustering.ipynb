```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
dataset = make_blobs(n_samples=200,
                    centers=4, n_features=2,
                    cluster_std=1.6, random_state=1)
```

```python
dataset
```

```python
df = dataset[0]
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(df)
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
cluster = kmeans.cluster_centers_
print(cluster)
```

```python
y_km = kmeans.fit_predict(df)
```

```python
plt.scatter(df[y_km == 0,0], df[y_km == 0,1], s=60, color='blue')
plt.scatter(df[y_km == 1,0], df[y_km == 1,1], s=60, color='red')
plt.scatter(df[y_km == 2,0], df[y_km == 2,1], s=60, color='green')
plt.scatter(df[y_km == 3,0], df[y_km == 3,1], s=60, color='orange')

plt.scatter(cluster[0][0], cluster[0][1], marker='*', s=200, color='black')
plt.scatter(cluster[1][0], cluster[1][1], marker='*', s=200, color='black')
plt.scatter(cluster[2][0], cluster[2][1], marker='*', s=200, color='black')
plt.scatter(cluster[3][0], cluster[3][1], marker='*', s=200, color='black')

plt.show()
```
